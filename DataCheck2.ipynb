```python
df = get_data_index('r1')
exp = 'RepFR2'
# sub = 'R1582E'
df_exp = df[df['experiment']==exp]
# df_sub = df[df['subject']==sub]
subjects = df['subject'].unique()
df_exp[['subject', 'session', 'experiment']]
# subjects
```

```python
subject = 'R1587J'
exp = 'RepFR2'
df_select = df[(df['subject']==subject) & (df['experiment']==exp)]
sessions = list(df[(df['subject']==subject) & (df['experiment']==exp)]['session'])
mont = int(df[(df['subject']==subject) & (df['experiment']==exp)].iloc()[0]['montage'])      #note that *usually* mont and loc will be 0.
loc = int(df[(df['subject']==subject) & (df['experiment']==exp)].iloc()[0]['localization'])
```
