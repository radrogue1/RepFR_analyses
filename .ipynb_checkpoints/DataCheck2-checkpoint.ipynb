```python
from cmlreaders import CMLReader, get_data_index
import numpy as np
import pandas as pd
```

```python
df = get_data_index('r1')
exp = 'RepFR2'
# sub = 'R1582E'
df_exp = df[df['experiment']==exp]
# df_sub = df[df['subject']==sub]
subjects = df['subject'].unique()
df_exp[['subject', 'session', 'experiment']].tail()
# subjects
```

```python
subject = 'R1587J'
exp = 'RepFR2'
df_select = df[(df['subject']==subject) & (df['experiment']==exp)]
sessions = list(df[(df['subject']==subject) & (df['experiment']==exp)]['session'])
mont = int(df[(df['subject']==subject) & (df['experiment']==exp)].iloc()[0]['montage'])      #note that *usually* mont and loc will be 0.
loc = int(df[(df['subject']==subject) & (df['experiment']==exp)].iloc()[0]['localization'])
```

```python
reader = CMLReader(subject, exp, sessions[0], montage=mont, localization=loc)
evs = reader.load('task_events')
# contacts = reader.load('contacts')
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
# evs[evs.session==0]
# evs.groupby('list')['recalled'].mean()
display(len(evs))
evs.head()
```

```python
pd.set_option('display.max_rows', None)
pairs = reader.load('pairs')
contacts = reader.load('contacts')
display(len(contacts))
display(contacts.head())
display(len(pairs))
display(pairs.head())
```
